# **Penjelasan Persamaan Matematika yang Digunakan**

Berikut adalah penjelasan dan rumus dari beberapa metrik evaluasi yang digunakan dalam klasifikasi:

1. Akurasi (Accuracy)

Akurasi mengukur seberapa banyak prediksi model yang benar dibandingkan dengan total data.

TP (True Positive): Prediksi positif yang benar
TN (True Negative): Prediksi negatif yang benar
FP (False Positive): Prediksi positif yang salah
FN (False Negative): Prediksi negatif yang salah
Akurasi tinggi menunjukkan model bekerja dengan baik, tetapi kurang cocok jika dataset tidak seimbang.

image.png

2. Presisi (Precision)

Presisi mengukur ketepatan prediksi positif model, yaitu seberapa banyak prediksi positif yang benar dibandingkan dengan semua prediksi positif. Semakin tinggi presisi, semakin sedikit kesalahan dalam klasifikasi positif.

image.png

3. Recall

Recall mengukur seberapa banyak data positif yang benar-benar terdeteksi oleh model dibandingkan dengan total data positif yang ada. Recall tinggi berarti model mampu menangkap banyak data positif.

image.png

4. F1 Score

F1 Score adalah rata-rata harmonik antara presisi dan recall, berguna untuk mengatasi ketidakseimbangan data. Jika presisi dan recall seimbang, maka F1 Score akan tinggi.

image.png

5. F1 Squared

F1Â² Score adalah variasi dari F1 Score yang lebih menekankan pada recall dibanding presisi. Berguna untuk kasus di mana mendeteksi semua kasus positif lebih penting daripada menghindari kesalahan positif.

image.png

6. ROC (Receiver Operating Characteristic) Curve


ROC adalah kurva yang menggambarkan trade-off antara True Positive Rate (Recall) dan False Positive Rate (FPR) pada berbagai ambang batas keputusan. ROC yang lebih dekat ke sudut kiri atas lebih baik, karena memiliki recall tinggi dan FPR rendah.

image.png

7. AUC (Area Under Curve)

AUC mengukur luas di bawah kurva ROC, menunjukkan kemampuan model dalam membedakan antara kelas positif dan negatif.
