# Penjelasan Persamaan Matematika yang Digunakan

1. Mean Squared Error (MSE)

MSE mengukur rata-rata kuadrat selisih antara nilai sebenarnya dan nilai prediksi. Semakin kecil nilai MSE, semakin baik model dalam memprediksi.

image.png

2. Root Mean Squared Error (RMSE)

RMSE adalah akar dari MSE, yang mengembalikan kesalahan ke satuan yang sama dengan data asli. RMSE lebih sensitif terhadap kesalahan besar dibandingkan MSE.

image.png

3. R-squared


  R-squared mengukur seberapa baik model menjelaskan variasi dalam data. Nilainya berkisar antara 0 hingga 1, dengan 1 menunjukkan model yang sempurna.
